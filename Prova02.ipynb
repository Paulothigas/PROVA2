```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1): 
        min_idx = i
    
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]  
    return arr
```

##### O loop só precisa ir até a penúltima posição porque quando o penúltimo elemento estiver no lugar certo o último elemento já vai estar no lugar certo 
##### Precisamos que min_idx seja o primeiro candidato a ser trocado na sublista não ordenada, e este candidato é a posição atual i 

##### O loop interno (j) começa a procura a partir do índice i + 1 até o fim, busca só precisa ser feita na parte desordenada do array.
##### A comparação agora vê se o elemento atual (arr[j]) é menor do que o menor encontrado até agora, o código anterior estava comparando errado arr[j] apenas com arr[i], sem achar o menor de toda a sublista.

#####  Usamos uma linha para trocar os valores de arr[i] e arr[min_idx] a troca anterior apagava o valor do menor elemento, deixando dois elementos iguais no array.

```python
import random

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1): 
        min_idx = i
        
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i] 
    return arr


s = [random.randint(1, 100) for i in range(7)] 

print("Lista Aleatória:", s)

s = selection_sort(s)
print("Lista Ordenada (Selection Sort):", s)
```

##### Ele começa examinando a lista [57, 84, 25, 29, 57, 48, 72]. Em cada passagem, ele seleciona o menor elemento da parte restante da lista e troca de lugar com o primeiro elemento dessa sublista que  não está ordenada. na primeira vez, ele encontra o 25 e o move para a primeira posição, trocando-o com o 57. Na segunda vez, ele encontra o 29 na parte restante da lista e o move para a segunda posição, trocando-o com o 84. Ele repete esse processo de encontrar o mínimo e trocar uma única vez até que toda a lista seja varrida, resultando na ordem final.

```python
def find_max(lista):
    if not lista:
        return None
    maior_valor_atual = lista[0]

    for elemento in lista[1:]:
        
        if elemento > maior_valor_atual:
            maior_valor_atual = elemento
    
    return maior_valor_atual

lista = [57, 84, 25, 29, 57, 48, 72]
maior = find_max(lista)
print(f"A lista é: {lista}")
print(f"O maior valor é: {maior}")
```
